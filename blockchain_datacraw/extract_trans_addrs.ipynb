```python
"""从区块数据中提取每个区块内的交易数据，以及交易数据涉及到的所有地址"""
import json
import format_transaction as ft
import json


def load_trans(transdata):
    """
    将格式好的交易数据存入transactions.json文件；
    因为每个区块内的交易是一个列表，为了让所有区块的交易以字典结构存到一个文件，就需要一笔一笔写入文件
    """
    with open('transactions.json','a') as f:
        for line in transdata:
            f.write(json.dumps(line,ensure_ascii=False,indent=2 )) 
            f.write(',')

def load_addrs(address):
    """
    将提取的所有交易地址存入address.json文件；
    """
    #必须以'a'方式写，不然只会存入最后一个区块的交易涉及的地址
    with open('address.json','a') as f:
        f.write('{"address":')
#         for line in address:
#             if line: 
        f.write(json.dumps(address,ensure_ascii=False,indent=2 ) + '\n')
        f.write('},')

        
def getblockhash_height(url,height=''):
    """从抓取的区块高度和区块哈希文件中单独提取出哈希值和高度"""
   
    with open(url,'r') as f:
        lines = json.loads(f.read())
        block= []
        for line in reversed(lines[:-5]):
            if height:
                block.append(line['height'])
            else:
                block.append(line['hash'])
        return block

def main():
    #blockheight = 532365
    #blockhash = "000000000000000000317d0ec9a48c7debd7a4d5010c087850a47c5d9f605da9"
    #blocks_height = [height for height in range(532365,532425)]
    #blocks_height = sorted(blocks_height,reverse = True)
    blocks_hash= getblockhash_height('height_hash.json')
    blocks_height = getblockhash_height('height_hash.json',True)
    #print(blocks_hash)
    #print(blocks_height)

    while blocks_height:
        #pop()删除后还能使用
        blockheight = blocks_height.pop()
        blockhash = blocks_hash.pop()
    #    print(blockheight)
    #    print(blockhash)
        url = 'blockdata/blcokheight_' + str(blockheight) + '.json'
        print(url)
    # 根据flag来判断，提取的是交易信息还是地址信息
    #     transdata = ft.format_transaction(url,blockheight,blockhash,flag=0)
    #     print(transdata)
    #     load_trans(transdata)

        addrs = ft.format_transaction(url,blockheight,blockhash)
        print(len(addrs))
#         print(addrs)
        load_addrs(addrs)
    
main()        

# blockheight = 532365
# blockhash = "000000000000000000317d0ec9a48c7debd7a4d5010c087850a47c5d9f605da9"
# url = 'blockdata/blcokheight_' + str(blockheight) + '.json'
# print(url)

# addrs = ft.format_transaction(url,blockheight,blockhash)
# print(len(addrs))
# # print(addrs)
# load_addrs(addrs)
```
