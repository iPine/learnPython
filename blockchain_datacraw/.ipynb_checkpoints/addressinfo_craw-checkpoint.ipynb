```python
from bs4 import BeautifulSoup
import requests
import json
import format_addrinfo as fa

def getHTMLText(url):
    """将HTML解析成文本返回
    """
    try:
        r = requests.get(url, timeout = 30)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return ""

def get_addrslist():
    """只使用与展示交易相关的所有地址;
    提取address.json文件里的地址，存入列表待用
    """
    with open('address.json','r') as f:
        addrs = json.loads(f.read())
    #     print(addrs['addrs'])
        addrs_list = []
        for addr in addrs['addrs']:
            #剔除重复地址
            for address in addr['address']:
                if address and (address not in addrs_list):
                    addrs_list.append(address)
        return addrs_list


def addressinfo_craw(address):
    """爬取地址信息
    """
    #address = '1EVzaFkkNNXq6RJh2oywwJMn8JPiq8ikDi'
    url = 'https://blockchain.info/rawaddr/' + address
    #print(url)

    html = getHTMLText(url)
    

    print('开始抓取')
    with open(address+'.json','w') as f:
        try:
        #必须要这样格式化了，方便format_address()读取json文件
            f.write('{"addrs":[')
            for line in html:
                f.write(line.strip())
            f.write(']}')

        except Exception:
            print('发生错误')

    print('抓取成功')

    

            
def load_addrinfo(addrinfo_list):
    """
    把处理好的地址信息列表存入本地addressinfo.json文件
    """
    with open('addressinfo.json','a') as f:
        for line in addrinfo_list:
            f.write(json.dumps(line,ensure_ascii=False,indent=2 )) 
            f.write(',')
        
            
 
    
def main():
    addrs_new = get_addrslist()
    print("需要提取信息的地址总数：",len(addrs_new))
    # print("需要提取信息的地址：{}".format(addrs_new))

    while addrs_new:
        address = addrs_new.pop()
        print("当前正在处理的地址：",address)
        if len(address) == 34:
            print("还未处理的地址数量：",len(addrs_new))
            addressinfo_craw(address)
            addressinfo = fa.format_address(address)
            #print(addressinfo)
            load_addrinfo(addressinfo)
        else:
            print('该地址长度无效！')
            continue
```
