```python
import matplotlib.pyplot as plt
%matplotlib inline

# x_values = [1,2,3,4,5]
x_values = list(range(1,5001))
y_values = [value**3 for value in x_values]

plt.scatter(x_values,y_values,edgecolors='none',c=x_values,cmap=plt.cm.cividis,s=40)
plt.title('Cube Numbers',fontsize=24)
plt.xlabel('Value',fontsize=14)
plt.ylabel('Cube of Value',fontsize=14)

plt.tick_params(axis='both',which='both',labelsize=14)
v = [0,5000,1,125000000000]
#参数tight的含义：changes x and y axis limits such that all data is shown.
plt.axis('tight')

plt.show()


```
