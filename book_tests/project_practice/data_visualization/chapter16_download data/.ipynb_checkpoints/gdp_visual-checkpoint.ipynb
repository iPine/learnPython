```python
import json

import pygal_maps_world.maps
from pygal.style import RotateStyle as RS, LightColorizedStyle as LCS

from country_codes import get_country_code

filename = 'gdp_json.json'

with open(filename) as f:
    gdp_data = json.load(f)
    #print(gdp_data)
    
    cc_gdp = {}
    # 打印每个国家2010年的gdp
    for gdp_dict in gdp_data:
        if gdp_dict['Year'] == 2016: 
            country_name = gdp_dict['Country Name']
            #将字符串转换为数字值
            gdp = round(float(gdp_dict['Value']),2)
            print(country_name + ": " + str(gdp))
            
            code = get_country_code(country_name)
            if code:
                cc_gdp[code] = gdp
            
    #根据gdp值将所有国家分成3组
    cc_gdps_1,cc_gdps_2,cc_gdps_3 = {},{},{}
    for cc,gdp in cc_gdp.items():
        if gdp < 1000000000:
            cc_gdps_1[cc] = gdp
        elif gdp < 10000000000:
            cc_gdps_2[cc] = gdp
        else:
            cc_gdps_3[cc] = gdp
    #打印看下每组有多少个国家
    print(len(cc_gdps_1),len(cc_gdps_2),len(cc_gdps_3))
    
    #Pygal将根据指定的颜色为每组选择颜色。
    wm_style = RS('#993366',base_style=LCS)
    #创建了一个 Worldmap 实例
    wm = pygal_maps_world.maps.World(style=wm_style)
    wm.title = 'World GDP in 2016, by Country'
  
    #wm.add('2016',cc_gdp)

    wm.add('0-10m',cc_gdps_1)
    wm.add('10m-100m',cc_gdps_2)
    wm.add('>1bn',cc_gdps_3)
    
    wm.render_to_file('world_gdp.svg')
```
